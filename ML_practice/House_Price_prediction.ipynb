## Create a data set

```python
import pandas as pd
```

```python
data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 4, 4, 5],
    'Bathrooms': [1, 2, 3, 3, 4],
    'Price': [3000000, 4500000, 6000000, 7500000, 9000000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

## Seprate feature and target

```python
x = df.drop(columns= "Price")
y = df[["Price"]]
```

## Step 3: Train Mosel

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

## Step 4: Predict New House Price

```python
predict = model.predict([[2000,4,3]])
print(predict)
```

## Step 5: Check Model Score

```python
print(model.score(x,y))
```
