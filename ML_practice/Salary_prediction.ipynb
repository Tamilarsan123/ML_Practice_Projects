# Problem Statement

* Predict an employee's salary based on their years of experience.

## Step 1:Create Dataset

```python
import pandas as pd
```

```python
data = {
    'Experiance':[1,2,3,4,5,6],
    'Salary':[25000,30000,40000,50000,60000,70000]
}

df = pd.DataFrame(data)
df
```

## Step 2:Define Features and Target

```python
x = df[["Experiance"]]
y = df[["Salary"]]
```

## Step 3:Train the Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

## Step 4:Predict Salary

```python
salary = int(input("Enter a Experiance:"))
prediction = model.predict([[salary]])
print(prediction)
```

## Step 5:Check accuracy

```python
print(model.score(x,y))
```
